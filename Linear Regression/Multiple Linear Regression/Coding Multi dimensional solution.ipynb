```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D # for plotting a 3d scatter plot
import matplotlib.pyplot as plt
```

```python
# load the data
X = []
Y = []
for line in open('/home/launchyard/Desktop/lazy_programmer/machine_learning_examples/linear_regression_class/data_2d.csv'):
    x1, x2, y = line.split(',')
    # creating a biased term x0
    X.append([float(x1), float(x2), 1]) # 1 goes last
    Y.append(float(y))
```

```python
# turn X and Y into numpy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
# plotting the data (3d) plot since there are three dimensions now.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
plt.show()
```

```python
# calculate weights of our model
# Remember we are not going to invert X transpose x.
# We are just going to use numpy linear algebra solver.
# Just remember that the default multiplication symbol (*) in numpy does element by element operation.
# But we want matrix multiplication here. Use (np.dot()) for that purpose.
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat = np.dot(X, w) # remember the order here (matrix multiplication)
```

#### Compute Rsquared

```python
# Same as before
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print("R squared: {}".format(r2)) # which is a good r squared that means the model is good.
```
